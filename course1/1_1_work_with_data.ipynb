## Working with Data

Data is an essential ingredient in data science. Data comes from various sources, such as experiments, computer simulations, surveys, field observations, online platforms, human-computer interactions, etc. There is a whole branch of **statistics** dealing with data collection, known as **experimental design**. As a general rule, we can think of the process generating the data as stochastic, because there is technical and epistemic uncertainty. In other words, there are conceptual limitations veiling the details of the data. For all these reasons, we are recommended to interpret data in the context of models; data would speak with the use of models.

Over this course we will assume that we already have collected the data. Our data will also be clean and tidy, which is uncommon in the real world though. We will make these assumptions in order to focus on understanding the data and modeling techniques associated with Bayesian Statistics.

A very useful skill when analyzing data is knowing how to write code in a programming language, such as Python and R. Even if your data is very clean and tidy, coding will still be essential since modern Bayesian Statistics is done mostly through programming languages such as Python and R.
