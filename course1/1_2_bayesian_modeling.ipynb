## Bayesian Modeling Introduction

In this lecture we are going to discuss Bayesian modeling. Models are simplified descriptions of a given system or process that, for some reason, we are interested in. Those descriptions are deliberately created in order to capture the most relevant aspects of the system.

There are various kinds of models; we focus our scope on Bayesian models. Particularly, we can summarize the Bayesian modeling process using three steps:

* Given some data and some assumptions on how this data could have been generated, we can design a model by combining building blocks known as **probability distributions**. Most of the time these models can provide a crude approximation of the data, and that's all we need.

* We also use Bayes' theorem, a probabilistic update rule, to add data to our models and logically derive the update of assumptions.

* We measure the goodness of models by checking whether the models make sense according to criteria like data, our subject knowledge and sometimes by comparing several models.

Bayesian models are also known as **probabilistic models** because the building block of which are probability distributions. But why do we use probabilities? It is because probabilities are the correct tool to not only model the variables, but also quantify the uncertainty. While statistical world has more than one definition of probability, from a Bayesian perspective, we define probability is **a measure that quantifies the uncertainty level** of a statement. We will use the modeling approach that explicitly takes uncertainty into account throughout the course.
