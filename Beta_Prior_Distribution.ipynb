```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
params = [0.5, 1, 2, 3]
x = np.linspace(0, 1, 100)
f, ax = plt.subplots(len(params), len(params), sharex=True, 
                     sharey=True,
                     figsize=(12, 10), constrained_layout=True)
for i in range(4):
    for j in range(4):
        a = params[i]
        b = params[j]
        y = stats.beta(a, b).pdf(x)
        ax[i,j].plot(x, y)
        ax[i,j].plot(0, 0, label=" = {:2.1f}\n = {:2.1f}".format(a, 
                     b), alpha=0)
        ax[i,j].legend()
ax[1,0].set_yticks([])
ax[1,0].set_xticks([0, 0.5, 1])
f.text(0.5, -0.01, '', ha='center')
f.text(-0.03, 0.5, 'p()', va='center', rotation=0)
plt.suptitle("Choosing alpha and beta parameters for Beta prior")
f
```
